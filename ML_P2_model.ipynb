```
#Load Libraries
import pandas as pd
import numpy as np
import seaborn as sns
sns.set(style='white', context='notebook', palette='deep')
```

```
#Load Data
train_data = pd.read_csv("/content/drive/My Drive/ML/tcd-ml-1920-group-income-train.csv")
test_data = pd.read_csv("/content/drive/My Drive/ML/tcd-ml-1920-group-income-test.csv")
```

```
rename_cols = {"Total Yearly Income [EUR]":'Income',
               "Work Experience in Current Job [years]":'Work_Exp',
               "Yearly Income in addition to Salary (e.g. Rental Income)":'Additional_Income',
               "Body Height [cm]":'Body_Height'}

train_data = train_data.rename(columns=rename_cols)
test_data = test_data.rename(columns=rename_cols)
```

```
#cat values
#'Housing Situation',
train_data['Housing Situation'] = train_data['Housing Situation'].replace(['0',0.0,'nA'],'Unknown Housing Situation')

#'Satisfation with employer'
#train_data['Satisfation with employer'] = train_data['Satisfation with employer'].ffill(axis = 0)
train_data['Satisfation with employer'] = train_data['Satisfation with employer'].fillna('Unknown')

#'Gender'
#train_data['Gender'] = train_data['Gender'].replace(['0','f'],['male','female'])
#train_data['Gender'] = train_data['Gender'].ffill(axis = 0)
train_data['Gender'] = train_data['Gender'].replace(['0','f',np.nan],'unknown')

#'Country'
train_data['Country'] = train_data['Country'].replace(['0',0.0],'honduras')
train_data['Country'] = train_data['Country'].fillna('honduras')

#'Profession'
freq_Profession = train_data['Profession'].mode().iloc[0]
train_data['Profession'] = train_data['Profession'].fillna(freq_Profession)

#'University Degree'
#train_data['University Degree'] = train_data['University Degree'].replace(['0'],'Unknown Degree')
#train_data['University Degree'] = train_data['University Degree'].ffill(axis = 0)
train_data['University Degree'] = train_data['University Degree'].replace(['0',np.nan],'Unknown')

#'Hair Color'
#train_data['Hair Color'] = train_data['Hair Color'].replace(['0'],'Unknown')
#train_data['Hair Color'] = train_data['Hair Color'].ffill(axis = 0)
train_data['Hair Color'] = train_data['Hair Color'].replace(['0',np.nan],'Unknown')

#num values

#'Year of Record'
train_data['Year of Record'] = train_data['Year of Record'].fillna(round(train_data["Year of Record"].median()))

#'Work_Exp'
train_data['Work_Exp'] = train_data['Work_Exp'].replace(['#NUM!'],np.nan)
#train_data['Work_Exp'] = train_data['Work_Exp'].ffill(axis = 0)
train_data['Work_Exp'] = train_data['Work_Exp'].fillna(0)
train_data['Work_Exp'] = pd.to_numeric(train_data['Work_Exp'])

#'Additional_Income'
train_data['Additional_Income'] = train_data['Additional_Income'].str.replace("EUR","")
train_data['Additional_Income'] = pd.to_numeric(train_data['Additional_Income'])
```

```
#cat values
#'Housing Situation',
test_data['Housing Situation'] = test_data['Housing Situation'].replace(['0',0.0,'nA'],'Unknown Housing Situation')

#'Satisfation with employer'
#test_data['Satisfation with employer'] = test_data['Satisfation with employer'].ffill(axis = 0)
test_data['Satisfation with employer'] = test_data['Satisfation with employer'].fillna('Unknown')

#'Gender'
#test_data['Gender'] = test_data['Gender'].replace(['0','f'],['male','female'])
#test_data['Gender'] = test_data['Gender'].ffill(axis = 0)
test_data['Gender'] = test_data['Gender'].replace(['0','f',np.nan],'unknown')

#'Country'
test_data['Country'] = test_data['Country'].replace(['0',0.0],'honduras')
test_data['Country'] = test_data['Country'].fillna('honduras')

#'Profession'
freq_Profession = test_data['Profession'].mode().iloc[0]
test_data['Profession'] = test_data['Profession'].fillna(freq_Profession)

#'University Degree'
#test_data['University Degree'] = test_data['University Degree'].replace(['0'],'Unknown Degree')
#test_data['University Degree'] = test_data['University Degree'].ffill(axis = 0)
test_data['University Degree'] = test_data['University Degree'].replace(['0',np.nan],'Unknown')

#'Hair Color'
#test_data['Hair Color'] = test_data['Hair Color'].replace(['0'],'Unknown')
#test_data['Hair Color'] = test_data['Hair Color'].ffill(axis = 0)
test_data['Hair Color'] = test_data['Hair Color'].replace(['0',np.nan],'Unknown')

#num values

#'Year of Record'
test_data['Year of Record'] = test_data['Year of Record'].fillna(round(test_data["Year of Record"].median()))

#'Work_Exp'
test_data['Work_Exp'] = test_data['Work_Exp'].replace(['#NUM!'],np.nan)
#test_data['Work_Exp'] = test_data['Work_Exp'].ffill(axis = 0)
test_data['Work_Exp'] = test_data['Work_Exp'].fillna(0)
test_data['Work_Exp'] = pd.to_numeric(test_data['Work_Exp'])

#'Additional_Income'
test_data['Additional_Income'] = test_data['Additional_Income'].str.replace("EUR","")
test_data['Additional_Income'] = pd.to_numeric(test_data['Additional_Income'])
```

```
num_cols = train_data.select_dtypes(include=np.number).columns.tolist()
cat_cols = train_data.select_dtypes(include=np.object).columns.tolist()
```

```
train = train_data
test = test_data
train = train.drop(['Income','Instance'], axis = 1)
test = test.drop(['Income','Instance'], axis = 1)
```

```
#Target Encoding
import category_encoders as ce
encoder = ce.TargetEncoder(cols=cat_cols)
encoder.fit(train,train_data['Income'])

encoded_train_data = encoder.transform(train)
encoded_test_data = encoder.transform(test)
#again check
```

```
X_train = encoded_train_data
X_test = encoded_test_data
```

```
Y_train = train_data['Income']
#Y_train = np.log(Y_train)
```

```
#train and test split
from sklearn.model_selection import train_test_split
xTrain, xTest, yTrain, yTest = train_test_split(X_train, Y_train, test_size = 0.2, random_state = 123)
```

```
import xgboost
from xgboost import XGBRegressor
xgb = XGBRegressor(
    eval_metric = 'mae',
    n_estimators = 800,
    learning_rate = 0.05,
    max_depth = 6,
    colsample_bytree = 1,
    reg_alpha = 7,
    subsample = 0.8,
    num_round = 20,
    eta = 1,
    gamma = 1
    )
```

```
#xgb.fit(X_train,Y_train,verbose=True)
eval_set = [(X_train, Y_train), (xTest, yTest)]
eval_metric = ["mae","error"]
%time xgb.fit(X_train, Y_train, eval_metric=eval_metric, eval_set=eval_set, verbose=True)
```

```
#Making predictions
print("Starting prediction")
prediction = xgb.predict(X_test)
#prediction = np.exp(prediction)
```

```
def positive_if_negative(x):
    if x < 0:
        return ((-1)*x)
    return x
   
y_pred2 = [positive_if_negative(x) for x in prediction]
```

```
sub_df = pd.DataFrame({'Instance':test_data['Instance'],
                       'Total Yearly Income [EUR]':y_pred2})
sub_df.head(20)
```
